```python
import tqdm
import pandas as pd
from tqdm import tqdm
```

```python
_synonym_cache = {}

def get_synonyms(word: str) -> list:
    """Fetch synonyms from WordNet with caching."""
    if word in _synonym_cache:
        return _synonym_cache[word]
    synonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            candidate = lemma.name().replace('_', ' ').replace('-', ' ').lower()
            if candidate != word.lower():          
                synonyms.add(candidate)
    result = list(synonyms)
    _synonym_cache[word] = result
    return result


def synonym_replacement(sentence: str, replace_prob: float = 0.4) -> str:

    words = sentence.split()
    new_words = []

    for word in words:

        if len(word) <= 2 or word == "":          
            new_words.append(word)
            continue

        synonyms = get_synonyms(word)

        if synonyms and random.random() < replace_prob:
            new_words.append(random.choice(synonyms))   
        else:
            new_words.append(word)                       

    return " ".join(new_words)


def augment_sentence_wn(sentence: str, num_augmented: int = 3, replace_prob: float = 0.4) -> list:
    
    return [
        synonym_replacement(sentence, replace_prob)
        for _ in range(num_augmented)
    ]


# ── Loop over full training DataFrame ────────────────────────────────────────

def apply_wn_augmentation(df: pd.DataFrame, text_col: str = "clean_text", label_col: str = "severity_class", target_classes: list = None, num_augmented: int = 3,replace_prob: float = 0.4,random_seed: int = 42) -> pd.DataFrame:
    
    random.seed(random_seed)

    if target_classes is not None:
        df_to_aug = df[df[label_col].isin(target_classes)].copy()
        df_no_aug = df[~df[label_col].isin(target_classes)].copy()
        print(f"Augmenting classes: {target_classes} "
              f"({len(df_to_aug)} samples) | Skipping: {len(df_no_aug)} samples")
    else:
        df_to_aug = df.copy()
        df_no_aug = pd.DataFrame(columns=df.columns)

    augmented_rows = []

    for _, row in tqdm(df_to_aug.iterrows(), total=len(df_to_aug), desc="WN Augmentation"):
        original_text = str(row[text_col])
        aug_texts = augment_sentence_wn(
            original_text,
            num_augmented=num_augmented,
            replace_prob=replace_prob
        )
        for aug_text in aug_texts:
            new_row = row.copy()
            new_row[text_col] = aug_text
            new_row["augmented"] = True
            augmented_rows.append(new_row)

    df_original = df.copy()
    df_original["augmented"] = False

    if augmented_rows:
        df_aug = pd.DataFrame(augmented_rows)
        df_combined = pd.concat([df_original, df_aug], ignore_index=True)
    else:
        df_combined = df_original

    df_combined = df_combined.sample(frac=1, random_state=random_seed).reset_index(drop=True)

    print(f"\n✅ WN Augmentation complete.")
    print(f"   Original : {len(df_original)}")
    print(f"   Generated: {len(augmented_rows)}")
    print(f"   Total    : {len(df_combined)}")
    print(f"\nNew class distribution:")
    print(df_combined[label_col].value_counts().sort_index())

    return df_combined
```
